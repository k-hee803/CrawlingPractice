```python
import sys
import os

import pandas as pd
import numpy as np

from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import chromedriver_autoinstaller

import time
from tqdm import tqdm_notebook

import warnings
warnings.filterwarnings('ignore')
```

```python
# 네이버 웹툰 창 열기
import chromedriver_autoinstaller
chrome_path = chromedriver_autoinstaller.install()
driver = webdriver.Chrome(chrome_path)

driver.get('https://comic.naver.com/webtoon/weekday')
time.sleep(2)
```

```python
#url 수집
url_list=[]
title_list=[]

articles = ".title"
article_raw = driver.find_elements_by_css_selector(articles)
article_raw[0]
```

```python
article_raw[0].get_attribute('href')
```

```python
for article in article_raw:
    url=article.get_attribute('href')
    url_list.append(url)
time.sleep(1)

for article in article_raw:
    title=article.text
    title_list.append(title)

    print(title)

print(len(url_list))
print(len(title_list))
```

```python
df=pd.DataFrame({'url':url_list, 'title':title_list})
df
```

```python
df.to_csv("webtoon_url.csv",encoding='utf-8-sig')
```

```python
dict = {}

number = 30

for i in tqdm_notebook(range(0,number)):
    url=url_load['url'][i]
    chrome_path = chromedriver_autoinstaller.install()
    driver = webdriver.Chrome(chrome_path)
    
    driver.get(url)
    time.sleep(1)
    
    target_info={}
    
    #제목
    overlays = ".title"
    tit = driver.find_element_by_css_selector(overlays)
    title = tit.text
    title
    
    #작가
    overlays = ".wrt_nm"
    wrt = driver.find_element_by_css_selector(overlays)
    writer = wrt.text
    writer
    
    #소개글
    st = driver.find_element_by_xpath('//*[@id="content"]/div/div[2]/p[1]')
    story = st.text
    story
    
    target_info['title'] = title
    target_info['writer'] = writer
    target_info['story'] = story
    
    dict[i] = target_info
    time.sleep(1)
    
    print(i, title)
    
    driver.close()
    
    result_df=pd.DataFrame.from_dict(dict,'index')
    
    result_df.to_csv("webtoon_content.csv",encoding='utf-8-sig')
    time.sleep(3)
```
